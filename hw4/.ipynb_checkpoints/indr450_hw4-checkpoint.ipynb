```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
import warnings
warnings.filterwarnings("ignore")
plt.style.use('ggplot')
```

```python
def MAE(sales, forecast):
    error = sales - forecast
    return np.mean(np.abs(error))
def MAPE(sales, forecast):
    error = sales - forecast
    return np.mean(np.abs(error / sales))
def RMSE(sales, forecast):
    error = sales - forecast
    return np.sqrt(np.mean(np.square(error)))
```

```python
audi = pd.read_csv("audi.csv")
```

```python
fig, ax = plt.subplots(figsize= (10,6), dpi = 200)
ax.plot(audi['Month'], audi['Sales'], label='real sales')
ax.set_xticks([12*i for i in range(int(len(audi)/12) + 1)])
ax.legend()
plt.show()
```

```python
audi_train = audi[4:84]
audi_test = audi[84:]
```

```python
x_train = audi_train.iloc[:,2:]
x_test = audi_test.iloc[:,2:]
y_train = audi_train.iloc[:,1]
y_test = audi_test.iloc[:,1]
x_train = sm.add_constant(x_train)
x_test = sm.add_constant(x_test)
```

# Question 1

```python
full_model = sm.OLS(y_train, x_train)
result1 = full_model.fit()
result1.summary()
```

```python
print("RMSE of full model on test set: {:.2f}".format(RMSE(result1.predict(x_test), y_test)))
print("MAPE of full model on test set: {:.2f}".format(MAPE(result1.predict(x_test), y_test)))
print("MAE of full model on test set: {:.2f}".format(MAE(result1.predict(x_test), y_test)))
```

```python
def lasso(alpha):
    lasso_result = Lasso(normalize=True)
    lasso_result.set_params(alpha = alpha)
    lasso_result.fit(x_train, y_train)  
    coefs = lasso_result.coef_
    predictors = np.array((np.where(coefs != 0)[0]))
    return predictors, RMSE(lasso_result.predict(x_test), y_test)
```

```python
alphas = np.linspace(0.1, 20, 200)
min_error = float('inf')
for alpha in alphas:
    predictors, rmse = lasso(alpha)
    if rmse < min_error:
        cur_predictors = predictors
        alpha_star = alpha
        min_error = rmse
x_train_reduced1 = x_train.iloc[:, cur_predictors]
x_train_reduced1 = sm.add_constant(x_train_reduced1)
x_test_reduced1 = x_test.iloc[:, cur_predictors]
x_test_reduced1 = sm.add_constant(x_test_reduced1)
```

```python
lassocv = LassoCV(alphas=np.linspace(0.1, 20, 200), cv=10, max_iter=10000, normalize=True)
lassocv.fit(x_train, y_train)
cv_predictors, cv_rmse = lasso(lassocv.alpha_)
x_train_reduced2 = x_train.iloc[:, cv_predictors]
x_train_reduced2 = sm.add_constant(x_train_reduced2)
x_test_reduced2 = x_test.iloc[:, cv_predictors]
x_test_reduced2 = sm.add_constant(x_test_reduced2)
```

```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, BaggingRegressor
```

```python
def regression_tree(depth):
    regression_tree = DecisionTreeRegressor(random_state = 23, max_depth=depth)
    result2 = regression_tree.fit(x_train, y_train)
    return result2, RMSE(result2.predict(x_train), y_train), RMSE(result2.predict(x_test), y_test)
```

```python
min_error = float('inf')
result2 = None
for depth in [2,3,4,5,8,9,10,11,12,15,20,30]:
    result, train_error, test_error = regression_tree(depth)
    if test_error < min_error:
        min_error = test_error
        result2 = result
```

```python
print("Train RMSE of the regression tree: {:.2f}".format(RMSE(result2.predict(x_train), y_train)))
print("Train MSE of the regression tree: {:.2f}".format(RMSE(result2.predict(x_train), y_train)**2))
print("Test RMSE of the regression tree: {:.2f}".format(RMSE(result2.predict(x_test), y_test)))
print("Test MSE of the regression tree: {:.2f}".format(RMSE(result2.predict(x_test), y_test)**2))
```

```python
bagged_tree = BaggingRegressor(random_state = 23, bootstrap_features = True)
result3 = bagged_tree.fit(x_train, y_train)
```

```python
print("Train RMSE of the bagged regression tree: {:.2f}".format(RMSE(result3.predict(x_train), y_train)))
print("Train MSE of the bagged regression tree: {:.2f}".format(RMSE(result3.predict(x_train), y_train)**2))
print("Test RMSE of the bagged regression tree: {:.2f}".format(RMSE(result3.predict(x_test), y_test)))
print("Test MSE of the bagged regression tree: {:.2f}".format(RMSE(result3.predict(x_test), y_test)**2))
```

```python
def random_forest(max_features):
    random_forest = RandomForestRegressor(random_state=23, max_features=max_features)
    result4 = random_forest.fit(x_train, y_train)
    return result4,RMSE(result4.predict(x_train), y_train), RMSE(result4.predict(x_test), y_test)
```

```python
min_error = float('inf')
result4 = None
for max_features in range(len(x_train.columns)):
    result, train_error, test_error = random_forest(max_features+1)
    if test_error < min_error:
        result4 = result
        min_error = test_error
```

```python
print("Train RMSE of the random forest: {:.2f}".format(RMSE(result4.predict(x_train), y_train)))
print("Train MSE of the random forest: {:.2f}".format(RMSE(result4.predict(x_train), y_train)**2))
print("Test RMSE of the random forest: {:.2f}".format(RMSE(result4.predict(x_test), y_test)))
print("Test MSE of the random forest: {:.2f}".format(RMSE(result4.predict(x_test), y_test)**2))
```

```python
Importance = pd.DataFrame({'Importance':result4.feature_importances_*100}, 
                          )

Importance = Importance.sort_values(by = 'Importance', 
                       axis = 0, 
                       ascending = True)

Importance.plot(kind = 'barh', 
                color = 'r')
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
def boosted_tree(depth, lr):
    boosted_tree = GradientBoostingRegressor(learning_rate=lr, max_depth=depth, random_state = 23)
    result5 = boosted_tree.fit(x_train, y_train)
    return result5, RMSE(result5.predict(x_train), y_train), RMSE(result5.predict(x_test), y_test)
```

```python
min_error = float('inf')
result5 = None
for lr in [0.1*i for i in range(1,11)]:
    for depth in [2,3,5,10,15,20]:
        result, train_error, test_error = boosted_tree(depth, lr)
        if test_error < min_error:
            min_error = test_error
            result5 = result
```

```python
print("Train RMSE of the boosted tree: {:.2f}".format(RMSE(result5.predict(x_train), y_train)))
print("Train MSE of the boosted tree: {:.2f}".format(RMSE(result5.predict(x_train), y_train)**2))
print("Test RMSE of the boosted tree: {:.2f}".format(RMSE(result5.predict(x_test), y_test)))
print("Test MSE of the boosted tree: {:.2f}".format(RMSE(result5.predict(x_test), y_test)**2))
```

```python
feature_importance = result5.feature_importances_*100

rel_imp = pd.Series(feature_importance, 
                    ).sort_values(inplace = False)

rel_imp.T.plot(kind = 'barh', 
               color = 'r', )

plt.xlabel('Variable Importance')

plt.gca().legend_ = None
```
