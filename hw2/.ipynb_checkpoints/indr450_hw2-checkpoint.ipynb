```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
plt.style.use('ggplot')
# import the random module
import random
import warnings
warnings.filterwarnings("ignore") # specify to ignore warning messages
import itertools
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
def MAE(sales, forecast):
    error = sales - forecast
    return np.mean(np.abs(error))
def MAPE(sales, forecast):
    error = sales - forecast
    return np.mean(np.abs(error / sales))
def RMSE(sales, forecast):
    error = sales - forecast
    return np.sqrt(np.mean(np.square(error)))
```

<h1> Question 1 </h1>

```python
coffee = pd.read_csv("CoffeePrice.csv", header = None, names = ["month", "price"])
coffee_price = coffee.price
coffee_month = coffee.month
```

```python
fig, ax = plt.subplots(figsize= (10,6))
ax.plot(coffee_month, coffee_price, label = "Coffee Price Index")
ax.legend()
ax.set_xticks([12*i for i in range(int(len(coffee)/12) + 1)])
for label in ax.get_xticklabels():
    label.set_rotation(45)
plt.show()
```

```python
detrended_coffee = np.array([coffee_price[i] - coffee_price[i-1] for i in range(1,len(coffee))])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5), dpi = 200)

fig = plot_acf(detrended_coffee, lags=20, ax=axes[0])
fig = plot_pacf(detrended_coffee, lags=20, ax=axes[1])
```

```python
mod_arma1 = sm.tsa.statespace.SARIMAX(coffee_price, trend ='c', order=(2, 0, 0),
                                seasonal_order=(0, 0, 0, 0))
results_arma1 = mod_arma1.fit(disp=False) 
results_arma1.summary()
```

```python
print("MAE of AR(2) model: {:.5f}".format(MAE(results_arma1.fittedvalues, coffee_price)))
print("MAPE of AR(2) model: {:.5f}".format(MAPE(results_arma1.fittedvalues, coffee_price)))
print("RMSE of AR(2) model: {:.5f}".format(RMSE(results_arma1.fittedvalues, coffee_price)))
```

```python
mod_arma2 = sm.tsa.statespace.SARIMAX(coffee_price, trend ='c', order=(1, 0, [0,1]),
                                seasonal_order=(0, 0, 0, 0))
results_arma2 = mod_arma2.fit(disp=False) 
results_arma2.summary()
```

```python
print("MAE of AR(1) & MA(1) model: {:.5f}".format(MAE(results_arma2.fittedvalues[1:], coffee_price[1:])))
print("MAPE of AR(1) & MA(1) model: {:.5f}".format(MAPE(results_arma2.fittedvalues[1:], coffee_price[1:])))
print("RMSE of AR(1) & MA(1) model: {:.5f}".format(RMSE(results_arma2.fittedvalues[1:], coffee_price[1:])))
```

```python
training_set = coffee_price[:int(0.75*len(coffee))]
test_set = coffee_price[int(0.75*len(coffee)):]
```

```python
mod_arma1 = sm.tsa.statespace.SARIMAX(training_set, trend ='c', order=(2, 0, 0),
                                seasonal_order=(0, 0, 0, 0))
results_arma1 = mod_arma1.fit(disp=False) 
test_result1 = results_arma1.predict(int(0.75*len(coffee)), len(coffee)-1)
```

```python
print("MAE of AR(2) model on test set: {:.5f}".format(MAE(test_set, test_result1)))
print("MAPE of AR(2) model on test set: {:.5f}".format(MAPE(test_set, test_result1)))
print("RMSE of AR(2) model on test set: {:.5f}".format(RMSE(test_set, test_result1)))
```

```python
mod_arma2 = sm.tsa.statespace.SARIMAX(training_set, trend ='c', order=(1, 0, [0,1]),
                                seasonal_order=(0, 0, 0, 0))
results_arma2 = mod_arma2.fit(disp=False) 
test_result2 = results_arma2.predict(int(0.75*len(coffee)), len(coffee)-1)
```

```python
print("MAE of AR(1) & MA(1) model on test set: {:.5f}".format(MAE(test_set, test_result2)))
print("MAPE of AR(1) & MA(1) model on test set: {:.5f}".format(MAPE(test_set, test_result2)))
print("RMSE of AR(1) & MA(1) model on test set: {:.5f}".format(RMSE(test_set, test_result2)))
```

<h1> Question 3 </h1>

```python
ausbeer['t'] = [i+1 for i in range(len(ausbeer))]
ausbeer['M1'] = [1,0,0,0] * int(len(ausbeer)/4) + [1,0]
ausbeer['M2'] = [0,1,0,0] * int(len(ausbeer)/4) + [0,1]
ausbeer['M3'] = [0,0,1,0] * int(len(ausbeer)/4) + [0,0]
```

```python
ausbeer
```

```python
model_beer_dummies = sm.OLS.from_formula('prod ~ M1 + M2 + M3', ausbeer)
result = model_beer_dummies.fit()
result.summary()
```

```python
model_beer_trend_dummies = sm.OLS.from_formula('prod ~ t + M1 + M2 + M3', ausbeer)
result = model_beer_trend_dummies.fit()
result.summary()
```
